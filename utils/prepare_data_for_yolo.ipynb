## Create Russia and Russia_oval annotations

#### Extract data from russia.csv

```python
russia = pd.DataFrame(pd.read_csv('Data/Russia/russian.csv'))
russia["filename"]='r' + russia["filename"]
russia
```

```python
Russia = pd.DataFrame(columns = russia.columns)
Russia_oval = pd.DataFrame(columns = russia.columns)
```

```python
path = 'Data/Russia_oval'

files = os.listdir(path)
for index, file in enumerate(files):
    ext = file.split('.')[-1]
    
    if ext in ['jpg', 'jpeg', 'png', 'JPG', 'JPEG', 'PNG']:
        rows = russia.loc[russia["filename"] == file]
        class_num = 0
        
        for i, row in rows.iterrows():
            
            annotation = ast.literal_eval(row['region_shape_attributes'])
            if not annotation:
                continue
            object_classes = ast.literal_eval(row['region_attributes'])
            if not object_classes:
                continue

            if object_classes['period'] == 'left' or object_classes['period'] == 'right':
                class_num = 3
            else:
                class_num = 0
                
            if object_classes['class'] == '0':
                row['region_id'] = 0
            if object_classes['class'] == '1':
                row['region_id'] = 1
                    
        if class_num == 3:
            Russia_oval = Russia_oval.append(rows)
        elif class_num == 0:
            Russia = Russia.append(rows)
```

## Rename images

```python
def rename_images(name):
    print(name)
    cls = name[:2]
    data = pd.DataFrame(pd.read_csv(f'Data/{name}/{name}.csv'))
#     data = data.drop(columns = ['Unnamed: 0'])
    
    new_df = pd.DataFrame(columns = data.columns)
    files = os.listdir(f'Data/{name}')
    for index, file in enumerate(files):
        
        ext = file.split('.')[-1]
        if ext in ['csv', 'ipynb_checkpoints']:
            continue
        new_name = randint(10000000, 100000000)
        
        os.rename(f'Data/{name}/{file}', f'Data/{name}/{new_name}{cls}.{ext}')
        data.loc[data['filename'] == file, 'filename'] = f'{new_name}{cls}.{ext}'
        
        data.to_csv(f'Data/{name}/{name}.csv', sep=',')
    print('saved')
    return data
```

```python
folders = ['Birmingham', 'Chester', 'Dublin', 'Edinburgh', 
           'Exeter', 'Germany', 'Glasgow', 'London', 
           'Newcastle', 'Russia', 'Russia_oval', 'Sheffield']

for folder in folders:
    rename_images(folder)
```

## Create txt annotations for each image

```python
def xywh_to_xyxy(x_min, y_min, box_width, box_height):
    x_max = x_min + box_width
    y_max = y_min + box_height
    return x_min, y_min, x_max, y_max

def xywh_to_xywh_scaled(x_min, y_min, box_height, box_width, image_width, image_height):
    x_center = x_min + box_width // 2
    y_center = y_min + box_height // 2
    return x_center / image_width, y_center / image_height, box_width / image_width, box_height / image_height
```

```python
def create_txt_annotations(name, group):
    
    data = pd.DataFrame(pd.read_csv(f'Data/{name}/{name}.csv'))
    
    for i, row in data.iterrows():
    
        annotation = ast.literal_eval(row['region_shape_attributes'])
        if not annotation:
            continue

        filename = row['filename'].split('.')[0]
        image = cv2.imread(f'Data/{name}/{row["filename"]}')

        width, height = None, None
        try:
            width, height = image.shape[1], image.shape[0]
        except AttributeError:
            print(row)
            pass

        x, y, w, h = xywh_to_xywh_scaled(annotation['x'], annotation['y'], annotation['width'], 
                                         annotation['height'], width, height)
    
        if group == 'british':
            class_num = row['region_id']
        elif group == 'german':
            class_num = 4
        elif group == 'russian':
            if row['region_id'] == 0:
                class_num = 0
            elif row['region_id'] == 1:
                class_num = 2
        elif group == 'russian_oval':
            if row['region_id'] == 0:
                class_num = 3
            elif row['region_id'] == 1:
                class_num = 2

        if os.path.exists(f'Data/{name}/{filename}.txt'):
            mode = 'a'
            line = f'\n{class_num} {x} {y} {w} {h}'
        else:
            mode = 'w'
            line = f'{class_num} {x} {y} {w} {h}'
        
        with open(f'Data/{name}/{filename}.txt', mode) as file:
            file.write(line)
        
    print(name, 'saved')
```

```python
british = ['Birmingham', 'Chester', 'Dublin', 'Edinburgh', 
           'Exeter', 'Glasgow', 'London', 'Newcastle', 'Sheffield']

for folder in british:
    create_txt_annotations(folder, 'british')
```

## Split into train and val

```python
from sklearn.model_selection import train_test_split
```

```python
def train_val(name):
    data = pd.DataFrame(pd.read_csv(f'Data/{name}/{name}.csv'))

    path = f'Data/{name}/'
    lst = list(data['filename'].unique())

    X = lst
    y = [f'{file.split(".")[0]}.txt' for file in X]
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)

    for img in X_train:
        shutil.copyfile(f'{path}{img}', f'Data/train/{img}')
    for text in y_train:
        shutil.copyfile(f'{path}{text}', f'Data/train/{text}')
    for img in X_val:
        shutil.copyfile(f'{path}{img}', f'Data/val/{img}')
    for text in y_val:
        shutil.copyfile(f'{path}{text}', f'Data/val/{text}')

    print(name, 'saved')
```

```python
folders = ['Germany', 'Birmingham', 'Chester', 'Dublin', 
           'Edinburgh', 'Exeter', 'Glasgow', 'London', 
           'Newcastle', 'Russia', 'Russia_oval', 'Sheffield']

for folder in folders:
    train_val(folder)
```
