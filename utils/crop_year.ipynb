```python
import cv2
import os
import pandas as pd
import ast
from random import randint
import shutil
```

```python
path = "Year/ImagesWithAnnotations"
files = os.listdir(path)

for index, file in enumerate(files):
    name, ext = '.'.join(file.split('.')[:-1]), file.split('.')[-1]
    if ext in ['jpg', 'png', 'jpeg']:
        text = name + '.txt'
        os.rename(f'{path}/{file}',f'{path}/{index:05}.jpg')
        os.rename(f'{path}/{text}',f'{path}/{index:05}.txt')
```

```python
path = "Year/ImagesWithAnnotations"
new_path = "Year/YearCropped"
files = os.listdir(path)

for index, file in enumerate(files):
    name, ext = '.'.join(file.split('.')[:-1]), file.split('.')[-1]
    if ext in ['jpg', 'png', 'jpeg']:
        text = name + '.txt'

        img = cv2.imread(path + '/' + file)
        with open(f'{path}/{text}', 'r') as f:
            count = 0
            for line in f:
                count += 1
                cls, x, y, w, h = list(map(float, line.split(' ')))
                if cls != 4: continue
                x, w = int(img.shape[1]*x), int(img.shape[1]*w)
                y, h = int(img.shape[0]*y), int(img.shape[0]*h)
                if h < 3 or w < 3: continue
                xmin, ymin, xmax, ymax = x-w/2, y-h/2, x+w/2, y+h/2
                cropped_img = img[int(ymin):int(ymax), int(xmin):int(xmax)]

                cv2.imwrite(f'{new_path}/{name}_{count}.jpg', cropped_img)
            print(f"cropped images for {file} are saved")
```
